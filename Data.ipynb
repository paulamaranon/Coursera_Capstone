##Data source The project will make use of the following data sources:

Postal codes of different neighbourhoods in Canada will be retrieved from this website https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M Location coordinates of different neighborhoods will be retrieved from this site: https://cocl.us/Geospatial_data

Finally, I will use the FourSquare API to explore different neighborhoods in Toronto. The Foursquare explore function will be used to get the most popular venues in each neighborhood.
